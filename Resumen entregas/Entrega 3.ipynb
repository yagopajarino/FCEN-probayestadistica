### Chebyshev

```python
def chebyshev(varianza,epsilon):
    return varianza / (epsilon ** 2)
```

```python
# Uso chebyshev
var = 0.1
epsilon = 0.5
chebyshev(var, epsilon)
```

### Ley Grandes Números

```python
def LGN(n, varianza, epsilon):
    return varianza / ((epsilon ** 2) * n)
```

```python
# Uso del LGN
n = 10
varianza = 0.1
epsilon = 0.5
LGN(n,varianza,epsilon)
```

### Raíces polinomios grado 2

```python
# Solver de raíces de polinomios de grado 2
def cuad(a,b,c):
    disc = b**2 - 4*a*c
    if (disc < 0):
        return "There are no real solutions"
    else:
        return ((-b + math.sqrt(disc))/(2*a), (-b - math.sqrt(disc))/(2*a))
```

```python
# Uso solver cuadráticas
a = 2
b = 1
c = 1
cuad(a,b,c)
```

### Teorema central del límite

```python
# Proba de Sn > x
def TCL_suma(esp, var, n, x):
    va = scipy.stats.norm()
    k = (x - esp * n)/(math.sqrt(var*n))
    return 1 - va.cdf(k)
```

```python
# Proba de Xn.raya > x
def TCL_promedio(esp, var, n, x):
    va = scipy.stats.norm()
    k = (x - esp)/(math.sqrt(var)/(math.sqrt(n)))
    return 1 - va.cdf(k)
```
